```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show

import hvplot.pandas
import plotly.express as px
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
#BRONX 

bronx_bachelors_csv = Path("Data/datausa_bronx_bachelor_concentration_2017.csv")
bronx_bachelor = pd.read_csv(bronx_bachelors_csv,parse_dates=True, infer_datetime_format=True)
bronx_bachelor_df = pd.DataFrame(bronx_bachelor)
#bronx_bachelor_df.head(10) 

bronx_bachelor_df2 = bronx_bachelor_df[['Year','Geography','CIP6','Completions']]
#bronx_bachelor_df2.head()

bronx_bachelor_df2 = bronx_bachelor_df2.rename(columns={"CIP6":"Field Of Study"})
bronx_bachelor_df2.head()
```

```python
bronx_top10_2017 = bronx_bachelor_df2.sort_values(ascending=False, by='Completions').head(10)
bronx_top10_2017
```

```python
#bronx_top10_2017.to_csv('bronx_top10_2017.csv')
```

```python
bronx_bach_pie = px.pie(bronx_top10_2017, values='Completions', names='Field Of Study',
             title='Bronx Top Ten Fields of Study for Bachelors Degree 2017',
             hover_data=['Field Of Study']) #labels={'lifeExp':'life expectancy'})
#fig.update_traces(textposition='inside', textinfo='percent+label')
bronx_bach_pie.write_image("../Images/bronx_bach_pie.png")
bronx_bach_pie.show()
```

```python
bronx_top_field_2017 = bronx_top10_2017.head(1)
```
